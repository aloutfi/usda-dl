## Final Formatting and column creation for our dataset!

```python
import pandas as pd
```

```python
df = pd.read_csv('data/polity_cultivar_pomo_usda.csv')
df.head()
```

```python
def afn(_afn):
    _afn = _afn.artist_full_name
    if isinstance(_afn, str):
        _afn = _afn.split(',')
        _afn = [x.strip() for x in _afn]
        _afn = _afn[0:2]
        _afn = ''.join(_afn)
    return _afn
df['artist_full_name'] = df.apply(lambda x: afn(x), axis=1)
```

```python
df.head()
```

```python
def foc(_x):
    _x = _x['geographic_origin']
    if isinstance(_x, str):
        _x = _x.split(',')[-1]
        _x = _x.strip()
    return _x
df['fruit_origin_country'] = df.apply(lambda x: foc(x), axis=1)
df.head()
```

```python
from collections import Counter
Counter(list(df['fruit_origin_country']))
```

```python
def fop(_x):
    _x = _x['geographic_origin']
    if isinstance(_x, str):
        try:
            _x = _x.split(',')[-2]
        except:
            _x = None
    return _x
df['fruit_origin_polity'] = df.apply(lambda x: fop(x), axis=1)
df.head()
```

```python
def iurl(_x):
    _x = _x['pomological_Id']
    _x = f'/{_x}.jpg'
    return _x
df['image_url'] = df.apply(lambda x: iurl(x), axis=1)
df.head()
```

```python
df['is_usda_recognized_plant_family']=df['is_usda_recognized']
final = ['artist_full_name', 'cultivar_name', 'fruit_common_name', 'fruit_origin_country',
        'fruit_origin_polity', 'fruit_scientific_name', 'fruit_variety', 'image_created_year',
        'image_url', 'is_originpolity_largest_producer', 'is_usda_recognized_plant_family',
        'pomological_Id']
final_df = df[final]
final_df.index.name = '_index'
```
