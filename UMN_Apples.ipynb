```python
from bs4 import BeautifulSoup
import requests
```

```python
r = requests.get('https://mnhardy.umn.edu/varieties/fruit/apples/all-apple-varieties')
soup = BeautifulSoup(r.content, 'html.parser')
print(soup.prettify())
```

```python
varieties = soup.find_all('h3')
varieties
```
